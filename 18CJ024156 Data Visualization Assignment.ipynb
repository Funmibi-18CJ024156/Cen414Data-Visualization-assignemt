```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


data = pd.read_csv(r'C:\Users\x\Downloads\mydatafile.csv')
df = pd.DataFrame(data)
average = pd.read_csv(r'C:\Users\x\Downloads\average.csv')
stand_dev= []
variance =[]
IQR =[]
df2 = pd.DataFrame(average)
Years = [2015, 2016, 2017, 2018, 2019]
plt.style.use('fivethirtyeight')


# PLOTTING THE GRAPH FOR THE FIVE YEARS
for i in range(1, 6, 1):
   
    X = list(df.iloc[:, 0])
    Y = list(df.iloc[:, i])
    plt.figure(figsize=(15,25))
    plt.barh(X, Y, height=0.9,align='edge', color='red')
    plt.title(f"Percentage of Government (public) television institutions in {Years[i-1]}", fontweight='bold', fontsize= 20, style='italic', fontname= 'Times New Roman')
    plt.xlabel("Percentage(%)", fontsize= 15, fontweight='bold')
    plt.ylabel("Country", fontsize= 15, fontweight='bold')
    plt.xticks(fontsize=14, fontname= 'Times New Roman')
    plt.yticks(fontsize=14, fontname= 'Times New Roman')   
  
# Show the plot
    plt.show()
    print("\n\n\n\n\n")
    
    
    
#AVERAGE PER YEAR PLOTTING
X = list(df2.iloc[:, 0])
Y = list(df2.iloc[:, 1])
plt.figure(figsize=(15,25))
plt.bar(X, Y, width=0.9,align='edge' )
plt.title("Average Percentage of Government TV institutions over the years", fontweight='bold', fontsize= 20, style='italic', fontname= 'Times New Roman')
plt.xlabel("Year", fontsize= 15, fontweight='bold')
plt.ylabel("Average", fontsize= 15, fontweight='bold')
plt.xticks(fontsize=14, fontname= 'Times New Roman')
plt.yticks(fontsize=14, fontname= 'Times New Roman')
# Show the plot
plt.show()
print("\n\n\n\n\n")
    
    
#STANDARD DEVIATION PER YEAR PLOTTING
stand_dev.append(df['2015'].std())
stand_dev.append(df['2016'].std())
stand_dev.append(df['2017'].std())
stand_dev.append(df['2018'].std())
stand_dev.append(df['2019'].std())
X = Years
Y = stand_dev
plt.figure(figsize=(15,25))
plt.bar(X, Y, width=0.9,align='edge', color='blue' )
plt.title("Standard Deviation of the Percentage of Government TV institutions over the years", fontweight='bold', fontsize= 20, style='italic', fontname= 'Times New Roman')
plt.xlabel("Year", fontsize= 15, fontweight='bold')
plt.ylabel("Stand Dev", fontsize= 15, fontweight='bold')
plt.xticks(fontsize=14, fontname= 'Times New Roman')
plt.yticks(fontsize=14, fontname= 'Times New Roman')

#VARIANCE PER YEAR PLOTTING
variance.append(df['2015'].var())
variance.append(df['2016'].var())
variance.append(df['2017'].var())
variance.append(df['2018'].var())
variance.append(df['2019'].var())
X = Years
Y = variance
plt.figure(figsize=(15,25))
plt.bar(X, Y, width=0.9,align='edge', color='blue' )
plt.title("Variance of the Percentage of Government TV institutions over the years", fontweight='bold', fontsize= 20, style='italic', fontname= 'Times New Roman')
plt.xlabel("Year", fontsize= 15, fontweight='bold')
plt.ylabel("Variance", fontsize= 15, fontweight='bold')
plt.xticks(fontsize=14, fontname= 'Times New Roman')
plt.yticks(fontsize=14, fontname= 'Times New Roman')

#INTERQUARTILE RANGE PER YEAR PLOTTING

for i in range(1, 6, 1):
    Q1 = df.iloc[:, i].quantile(.25)
    Q3 = df.iloc[:, i].quantile(.75)
    IQR.append(Q3-Q1)
P = Years
Q = IQR
plt.figure(figsize=(15,25))
plt.bar(P, Q, width=0.9,align='edge', color='blue')
plt.title("Interquartile range of the Percentage of Government TV institutions over the years", fontweight='bold', fontsize= 20, style='italic', fontname= 'Times New Roman')
plt.xlabel("Year", fontsize= 15, fontweight='bold')
plt.ylabel("IQ Range", fontsize= 15, fontweight='bold')
plt.xticks(fontsize=14, fontname= 'Times New Roman')
plt.yticks(fontsize=14, fontname= 'Times New Roman')
  
# Show the plot
plt.show()
    
    



```
